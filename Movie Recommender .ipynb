```python
import ast

def convert(text):
    L = []
    if isinstance(text, str):
        try:
            items = ast.literal_eval(text)
        except:
            return []  # return empty list if parsing fails
    elif isinstance(text, list):
        items = text
    else:
        return []

    for i in items:
        # Handle if item is string (like 'action') or dict (like {'name': 'action'})
        if isinstance(i, dict) and 'name' in i:
            L.append(i['name'])
        elif isinstance(i, str):
            L.append(i)
    return L
```

```python

movies_merge['genres'] = movies['genres'].apply(convert)
movies_merge.head()
```

```python
movies_merge['keywords'] = movies_merge['keywords'].apply(convert)
movies_merge.head()
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
movies_merge['cast'] = movies_merge['cast'].apply(convert)
movies_merge.head()
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies_merge['crew'] = movies_merge['crew'].apply(fetch_director)
```

```python
movies_merge.sample(5)
```

```python

movies_merge['overview'] = movies_merge['overview'].apply(lambda x:x.split())
```

```python
movies_merge.head()
```

```python
def collapse(L):
    L1 = []
    if isinstance(L, float):  # means it's NaN
        return L1
    for i in L:
        if isinstance(i, str):  # ensure it's a string
            L1.append(i.replace(" ", ""))
    return L1
```

```python
movies_merge['cast'] = movies_merge['cast'].apply(collapse)
movies_merge['crew'] = movies_merge['crew'].apply(collapse)
movies_merge['genres'] = movies_merge['genres'].apply(collapse)
movies_merge['keywords'] = movies_merge['keywords'].apply(collapse)
```

```python
movies_merge.head()
```

```python

movies_merge['tags'] = movies_merge['overview'] + movies_merge['genres'] + movies_merge['keywords'] + movies_merge['cast'] + movies_merge['crew']
```

```python
movies_merge.head()
```

```python
new_movies_merge= movies_merge[['movie_id' , 'title'  , 'tags']]
```

```python
new_movies_merge['tags'] = new_movies_merge['tags'].apply(lambda x: " ".join(x))
new_movies_merge.head()
```

# vactorization

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
    
```

```python
vector = cv.fit_transform(new_movies_merge['tags']).toarray()
```

```python
vector
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity.shape
```

```python
def recommend(movie):
    # Find the index of the movie by title
    index = new_movies_merge[new_movies_merge['title'] == movie].index[0]

    # Get similarity scores for this movie
    distances = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda x: x[1])

    # Print top 5 recommended movie titles (excluding the input movie itself)
    for i in distances[1:6]:
        print(new_movies_merge.iloc[i[0]].title)
```

```python
recommend('Gandhi')
```
